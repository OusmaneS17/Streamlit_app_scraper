```python
import pandas as pd
```

```python
data = pd.read_csv('Brute/climatisation.csv')
```

```python
data.head(3)
```

# Nettoyage des colonnes

```python
# Supprimer la premiere ligne
data.drop(index=0, inplace=True)
```

```python
# Liste des colonnes à supprimer
colonnes_a_supprimer = ['web-scraper-order', 'web-scraper-start-url', 'url', 'url-href']

# Supprimer les colonnes
data.drop(columns=colonnes_a_supprimer, inplace=True)

# Afficher les premières lignes du DataFrame pour vérifier
data.head()
```

```python
# Traiter la colonne prix et adresse
# Nettoyer la colonne Prix
data['Prix'] = data['Prix'].str.replace(' F Cfa', '').str.replace(' ', '').astype(int)

# Nettoyer la colonne Adresse
data['Adresse'] = data['Adresse'].str.split(',').str[0]

# Afficher les premières lignes du DataFrame pour vérifier
data.head()
```

## 3. Sauvegarder le fichier nettoyé 

```python
data.to_excel('Cleaned/Climatisation_cleaned.xlsx', index=False)
```
